```python
# Cell 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Cell 2: Gender distribution dataset
gender_data = {'Gender': ['Male', 'Female', 'Other', 'Female', 'Male', 'Female', 'Male', 'Other', 'Female', 'Male']}
gender_df = pd.DataFrame(gender_data)

# Cell 3: Age distribution dataset
age_data = {'Age': [25, 45, 30, 35, 50, 23, 40, 70, 60, 28, 33, 48, 37, 20, 55, 27, 32, 43, 36, 38]}
age_df = pd.DataFrame(age_data)

# Cell 4: Bar Chart for Gender Distribution
gender_counts = gender_df['Gender'].value_counts()
plt.figure(figsize=(8, 6))
gender_counts.plot(kind='bar', color=['blue', 'pink', 'green'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()

# Cell 5: Histogram for Age Distribution
plt.figure(figsize=(10, 6))
plt.hist(age_df['Age'], bins=10, color='purple', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
